```python
# 4b)
df = pd.read_csv('test.csv')
```

```python
# 4c)Calculate similarity between description_x and description_y and store resultant scores in a new column
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

```python
a=[]
def similarity(df):
    for i in range(len(df)):
        a.append(similar(df["description_x"][i],df["description_y"][i]))
        df["Similarity"]=pd.Series(a)
    return df

similarity(df)
```
