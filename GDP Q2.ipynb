```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Sales_Transactions_Dataset_Weekly.csv')
```

```python
#2b) Identify the best performing product (based on volume)
df2_b=df.loc[:,'Product_Code':'W51']
def total (df2_b):
    for i in range(len(df2_b)):
        df2_b["SUM"]=df2_b.sum(axis=1)
        a=df2_b.sort_values('SUM',ascending=False)["Product_Code"].iloc[0]
        return a
total (df2_b)
```

```python
#2c) Identify the most promising product (emerging product)
df2_c=df.loc[:,'Product_Code':'W51']
def aa(df2_c):
    a=(df2_c == 0).sum(axis=1)
    df2_c["Count"]=pd.Series(a.values)
    df2_c=df2_c[df2_c["Count"]>20]
    for i in range(len(df2_c)):
        df2_c["SUM"]=df2_c.sum(axis=1)-df2_c["Count"]
        b=df2_c.sort_values('SUM',ascending=False)["Product_Code"].iloc[0]
        return b
    
aa(df2_c)
```

```python
#2d) Identify the worst performing product on a biweekly basis
df2_d=df.loc[:,'Product_Code':'W51']
df2_dNew=df2_d.groupby((np.arange(len(df2_d.columns)) // 2) + 1, axis=1).sum().add_prefix('s')
df2_dNew["Product Code"]=df['Product_Code']
df2_dNew["Minimum Bi-weekly"]=df2_dNew.max(axis=1)
df2_dNew.sort_values('Minimum Bi-weekly',ascending=True)["Product Code"].iloc[0]
```

```python
#2e) Identify outliers from the data and output the corresponding week numbers
cols = [x for x in range(55)]
df2_e=df.drop(df.columns[cols],axis=1)
wlst=df2_e.columns
for i in range(len(df2_e)):
    lst=list(df2_e.iloc[i,:])
    print('product: ','P',i+1)
    for j, w in zip(lst, wlst):
        if (j<0.05) | (j>0.95):
            print(j,'Week:',w[-2:])
    print()
```
